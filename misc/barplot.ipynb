# Bar Plot

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from scipy.misc import face
%matplotlib inline
```

```python
image = face()

probs = torch.Tensor([[27887.1387,   605.1686,   430.5393,   333.2011,   160.9669]])

classes =  ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat']
probs = probs / probs.sum()

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.bar(classes, probs.view(probs.shape[-1]), align='center')
plt.subplot(122)
plt.imshow(image)
```
