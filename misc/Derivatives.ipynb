```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)  # prepare x
y = x ** 3            # our function x^3
y_prime = 3 * x ** 2  # first order derivitive of x^3 is 3 * x^2
y_2prime = 6 * x      # derivative of 3*x^2 is 6*x (second derivative)

plt.figure(figsize=(8, 10))  # plot size
plt.plot(x, y, label='y', linewidth=3)  # plot y
plt.plot(x, y_prime, label="y'", linewidth=3)  # plot y'
plt.plot(x, y_2prime, label="y''", linewidth=3)  # plot y''

plt.xlim([-5.5, 5.5])  # set the limits from x-axis
plt.ylim((-25, 50))    # the smae for y-axis

plt.legend()  # add legend
```
