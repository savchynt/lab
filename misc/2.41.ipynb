```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x1 = np.arange(-1.5, 1.5, 0.1)
x2 = np.arange(-1.5, 1.5, 0.1)
x1, x2 = np.meshgrid(x1, x2)
s = sigmoid(x1 + x2)
s2 = sigmoid(10 * x1 + 10 * x2)
levels = np.arange(0, 1, 0.01)

plt.subplot(121)
plt.title('sigmoid(x1 + x2)')
c1 = plt.contourf(x1, x2, s, cmap=cm.RdBu, levels=levels)
plt.plot([-1], [-1], marker='o', markersize=10, color="red")
plt.plot([1], [1], marker='o', markersize=10, color="blue")
plt.colorbar(c1)

plt.subplot(122)
plt.title('sigmoid(10x1 + 10x2)')
c2 = plt.contourf(x1, x2, s2, cmap=cm.RdBu, levels=levels)
plt.plot([-1], [-1], marker='o', markersize=10, color="red")
plt.plot([1], [1], marker='o', markersize=10, color="blue")
plt.colorbar(c1)

plt.subplots_adjust(top=2, bottom=1, left=0, right=2)
```
