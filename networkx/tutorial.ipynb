```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def draw(G):
    nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G = nx.Graph()
draw(G)
```

```python
G.add_node(1)
draw(G)
```

```python
G.add_nodes_from([2, 3])
draw(G)
```

```python
H = nx.path_graph(10)
draw(H)
```

```python
G.add_node(H)
draw(G)
```

```python
G.add_edge(1, 2)
draw(G)
```

```python
G = nx.Graph()
e = (2, 3)
G.add_edge(*e)
draw(G)
```

```python
G = nx.Graph()
G.add_edges_from([(1, 2, {'weight': 10}), (1, 3)])
G.add_node(4)
draw(G)
```

```python
MG = nx.MultiGraph()
MG.add_weighted_edges_from([(1, 2, 0.5), (1, 2, 0.75), (2, 3, 0.5)])
draw(MG)
```
